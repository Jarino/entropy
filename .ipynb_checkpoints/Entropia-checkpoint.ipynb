# Experimenty s entropiou

The goal is to test entropy as a metric, which we could use as a fitness function, for our experiment with genetic programming and dimension reducion for scatter plot visualization of multivariate data.

What needs to be tested:
- what values will the entropy provide for two basic cases: random data a two clusters
- what will be the difference, when there will be three clustres

-----------------------

Cieľom je otestovať entropiu ako metriku, ktorú by sme použili ako fitness funkciu pre náš experiment s genetickým programovaním a redukciou dimenzií pre zobrazenie na scatter plote.

Čo treba otestovať:
- aké hodnoty entropia dáva pre dva základné prípady: náhodné dáta, roztrúsené kade tade a dva relatívne OK zhluky
- aký bude rozdiel, ak sú zhluky tri

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats

%matplotlib inline
```

Najprv si teda vygenerujeme dva datasety, jeden s dátami pripomínajúcimi šum a jeden s peknými zhlukmi.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))
plt.xlim(0,10)
plt.ylim(0,10)

# sumove data
noise_data = np.random.rand(100,2)*10

# zhluky dat
cluster_a = 0.5 * np.random.randn(50,2) + 2
cluster_b = 0.5 * np.random.randn(50,2) + 8
cluster_data = np.concatenate((cluster_a, cluster_b), axis=0)


ax1.scatter(noise_data[:,0], noise_data[:,1])
ax2.scatter(cluster_data[:,0], cluster_data[:,1])
```

Teraz vypočítame entropiu pre jednotlivé datasety:

```python
stats.entropy(noise_data)
```

```python
stats.entropy(cluster_data)
```

Takéto straightforward použitie entropy funkcie zrejme nefunguje. Neviem ako funguje pod kapotou, ale napadá ma, že predpokladá, že už do nej sypeme biny z histogramu, takže by sme mohli ešte vyskúšať najprv zistiť početnosť a nasypať tam to.

```python
noise_hist = np.histogram(noise_data)
noise_entr = stats.entropy(noise_hist[0]/100)
```

```python
clustered_hist = np.histogram(cluster_data)
clustered_entr = stats.entropy(clustered_hist[0]/100)
```

Toto vyzerá o niečo lepšie, skúsime dam našenkovať jeden zhluk, malo by to ísť bližšie k jednotke.

```python
one_cluster_data = 0.5 * np.random.randn(100,2) + 5
one_cluster_hist = np.histogram(one_cluster_data)
one_cluster_entr = stats.entropy(one_cluster_data[0]/100)

three_cluster_data = np.concatenate(
    (.1 * np.random.randn(33,2) + 1,
     .1 * np.random.randn(33,2) + 5,
     .1 * np.random.randn(34,2) + 9
    ), axis=0)

three_cluster_hist = np.histogram(three_cluster_data)
three_cluster_entr = stats.entropy(three_cluster_hist[0]/100)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(15,10))

ax1.scatter(noise_data[:,0], noise_data[:,1])
ax1.set_title('Entropy: %f'%(noise_entr))
ax1.set_xlim([0, 10])
ax1.set_ylim([0, 10])

ax2.scatter(one_cluster_data[:,0], one_cluster_data[:,1])
ax2.set_title('Entropy: %f'%(one_cluster_entr))
ax2.set_xlim([0, 10])
ax2.set_ylim([0, 10])

ax3.scatter(cluster_data[:,0], cluster_data[:,1])
ax3.set_title('Entropy: %f'%(clustered_entr))
ax3.set_xlim([0, 10])
ax3.set_ylim([0, 10])

ax4.scatter(three_cluster_data[:,0], three_cluster_data[:,1])
ax4.set_title('Entropy: %f'%(three_cluster_entr))
ax4.set_xlim([0, 10])
ax4.set_ylim([0, 10])
```
